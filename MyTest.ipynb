Таблиця з sample_binary під змінною ds_samplе

```python
ds_sample=pd.read_csv(r"C:\Users\admin\Desktop\Edvantis\sample_binary_Y.csv",sep="|", header=0, float_precision="round_trip", low_memory=False)
ds_sample.head(20)
                
```

Таблиця з «tax assessor» задана під змінною ds_tа

```python
ds_ta=pd.read_csv(r"C:\Users\admin\Desktop\Edvantis\ta_source_encrypted.csv",sep="|", header=0, float_precision="round_trip", low_memory=False)
ds_ta.head(10)
```

Джоїн таблиць sample_binary та ta_source_encrypted у join_keys_for_TA . Видалення дублюючих значень та сортування по attomId

```python
join_keys_for_TA = pd.merge(ds_sample[['attomId','sample_type']], ds_ta[['attom_id', 'fulladdress', 'PartyOwner1NameFirst', 'PartyOwner1NameLast', 'SitusCounty','EstimatedMinValue','EstimatedMaxValue','ConfidenceScore']], left_on='attomId', right_on='attom_id', how='inner')
del join_keys_for_TA['attom_id']

```

```python
join_keys_for_TA.sort_values(by=['attomId'], inplace=True)
join_keys_for_TA = join_keys_for_TA.drop_duplicates(subset = ["fulladdress"])
join_keys_for_TA
```

Джоїн таблиць join_keys_for_TA та nb_source_encrypted.

```python
join_keys_for_NB = pd.merge(join_keys_for_TA, ds[['fulladdress', 'personfirstname', 'personlastname','personexactage']], left_on=['fulladdress', 'PartyOwner1NameFirst', 'PartyOwner1NameLast'], right_on=['fulladdress', 'personfirstname', 'personlastname'], how='left')

join_keys_for_NB.head(5)
```

Видалення дублікатів

```python
join_keys_for_NB.drop_duplicates(subset='fulladdress', keep ='first',inplace=True )
join_keys_for_NB.head(5)
```

Утворення нової таблиці, конкатенацією

```python
fulldf=pd.concat([join_keys_for_NB,join_keys_for_TA])
fulldf.head(10)
```

Видаленны порожніх значень та значень які містять str, а не float

```python
df_clean=fulldf.drop(columns=['fulladdress', 'PartyOwner1NameFirst','PartyOwner1NameLast','personfirstname','personlastname','sample_type','SitusCounty'])

df_clean.dropna(inplace=True)
df_clean.head(10)
```

```python
from sklearn.model_selection import cross_val_score, train_test_split, cross_val_predict

from sklearn import metrics
X_train, X_test, y_train, y_test = train_test_split(df_clean,df_clean,test_size=.2)
```

```python
from sklearn.multioutput import MultiOutputClassifier

from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
knn_model = KNeighborsClassifier(n_neighbors=5,p=2)
knn_model.fit(X_train, y_train)
y_pred = knn_model.predict(X_test)
y_pred





```

```python
from sklearn.multioutput import MultiOutputClassifier
from sklearn.metrics import roc_auc_score, accuracy_score
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(max_depth=5, random_state=0)
clf.fit(X_train, y_train)
y_pred1 = clf.predict(X_test)

roc_auc_score(y_test, y_pred1)
```
